## Import Modules

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Binary Step Activation

```python
def binary_step_act(x):
    """
    Binary Step Activation Function
    Arguments:
    x : input data points 
    """
#     if x >= 0:
#         return 1
#     else:
#         return 0
    return(np.where(x>=0,1,0))
```

```python
z = np.linspace(-1,1,20)
z
```

```python
# Calculate binary step activation for input values
bin_step_out = [binary_step_act(x) for x in z]
```

```python
# Display plot
plt.title('Binary Step Activation Function')
plt.plot(z, bin_step_out, label='Binary Step Function')
plt.xlabel('z')
plt.ylabel('bin_step_out')
plt.legend()
plt.show()
```

## Sigmoid Activation Function

```python
def sigmoid(x):
    """
    sigmoid activation function
    """
    return 1.0/(1.0 + np.exp(-x))
```

```python
x_sig = np.linspace(-10,10,100)
x_sig
```

```python
sig_out = [sigmoid(x) for x in x_sig]
```

```python
# Display plot
plt.title('Sigmoid Activation Function')
plt.plot(x_sig, sig_out, label='Sigmoid Activation Function')
plt.xlabel('x_sig')
plt.ylabel('sig_out')
plt.legend()
plt.show()
```

## Tanh or Hyperbolic Tangent Activation Function

```python
# Numpy has inbuilt tanh function
tanh_out = [np.tanh(x) for x in x_sig]
```

```python
# Display plot
plt.title('tanh Activation Function')
plt.plot(x_sig, tanh_out, label='tanh Activation Function')
plt.xlabel('x_sig')
plt.ylabel('tanh_out')
plt.legend()
plt.show()
```

## ReLU Activation Function

```python
def relu(x):
    return np.where(x>=0,x,0)
```

```python
relu_out = [relu(x) for x in x_sig]
```

```python
# Display plot
plt.title('ReLU Activation Function')
plt.plot(x_sig, relu_out, label='ReLU Activation Function')
plt.xlabel('x_sig')
plt.ylabel('relu_out')
plt.legend()
plt.show()
```
